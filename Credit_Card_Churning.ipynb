# **Credit Card Churnüè¶**

>  **Business Problem:** A business manager of a consumer credit card bank is facing the problem of customer attrition. They want to analyze the data to find out the reason behind this and leverage the same to predict customers who are likely to drop off.üí≥

> **Credit card churning:** is the process of opening cards for the sole purpose of earning welcome bonuses or other benefits. Usually, it involves closing cards after the bonus posts to your account and before the next annual fee is charged. A pattern of opening and closing cards quickly is often a red flag for card issuers.ü§ë

> **Columns description**: [Credit Card Churn](https://www.kaggle.com/datasets/anwarsan/credit-card-bank-churn)
* CLIENTNUM: Unique identifier for the customer holding the account
* Attrition_Flag: Existing Customer	= 1 , Attrited Customer = 0
* Customer_Age: Customer's Age in Years
* Gender: M=Male, F=Female
* Dependent_count: Number of dependents
* Education_Level: Educational Qualification of the account holder (example: high school, college graduate, etc.)
* Marital_Status: Married, Single, Divorced, Unknown
* Income_Category: Annual Income Category of the account holder (< $40K, $40K - 60K, $60K - $80K, $80K-$120K, > $120K, Unknown)
* Card_Category: Type of Card (Blue, Silver, Gold, Platinum)
* Months_on_book: Period of relationship with bank
* Total_Relationship_Count: total number of products held by the customers (cards, accounts, etc.)
* Months_Inactive_12_mon: No. of months inactive in the last 12 months
* Contacts_Count_12_mon: No. of Contacts in the last 12 months
* Credit_Limit: Credit Limit on the Credit Card (A credit limit is the maximum amount you can charge on a revolving credit account, such as a credit card.)
* Total_Revolving_Bal: Total Revolving Balance on the Credit Card (If you don't pay the balance on your revolving credit account in full every month, the unpaid portion carries over to the next month).
* Avg_Open_To_Buy: Open to Buy Credit Line (Average of last 12 months)
* Total_Amt_Chng_Q4_Q1: represents how much the customer increased their expenditure when comparing the 4th quarter against the 1st (Q4 over Q1).
* Total_Trans_Amt: Total Transaction Amount (Last 12 months)
* Total_Trans_Ct: Total Transaction Count (Last 12 months)
* Total_Ct_Chng_Q4_Q1: Change in Transaction Count (Q4 over Q1)
* Avg_Utilization_Ratio: is the amount of revolving credit you're currently using divided by the total amount of revolving credit you have available. In other words, it's how much you currently owe divided by your credit limit. It is generally expressed as a percent.

## **DATA PREPROCESSING‚öôÔ∏è**


 ### Encoding Variables + Set features and target columns

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
data = pd.get_dummies(data, columns=['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']) #covert category columns
le = LabelEncoder()
y = data['Attrition_Flag']
y = le.fit_transform(y)
x = data.iloc[:,data.columns != 'Attrition_Flag'] #take the features which is from column index 1 to the last column
```

```python
pd.DataFrame(x)
```

```python
pd.DataFrame(y)
```

 ### Split the dataset for training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0) #random_state=0 to get the same train and test sets across different executions
```

## **MODEL BUILDINGüß±**

### 1. Model One: Decision Tree Classification Algorithmüå≥

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree = DecisionTreeClassifier() #create object
tree.fit(x_train, y_train) #train the object

tree_pred = tree.predict(x_test) #predict for the unlabeled data
tree_accuracy = accuracy_score(y_test, tree_pred) #calculate the accuracy between unlabel data and actual data

print("Accuracy Score of Tree Classification Algorithm is: {:0.2f}%".format(tree_accuracy*100))
```

### 2. Model two: Logistic Regression Algorithm

```python
from sklearn.linear_model import LogisticRegression

regression = LogisticRegression()
regression.fit(x_train, y_train)

regression_pred = regression.predict(x_test)
regression_accuracy = accuracy_score(y_test, regression_pred)

print("Accuracy Score of Logistic Regression Algorithm is: {:0.2f}%".format(regression_accuracy*100))
```

```python
from sklearn.metrics import plot_confusion_matrix
fig, ax = plt.subplots(figsize=(7,7))
plot_confusion_matrix(regression, x_test, y_test, cmap=plt.cm.Blues, display_labels = ['Existing','Attrited'], ax=ax)
plt.title('Confusion Matrix', fontsize=20)
plt.show()
```

### 3. Model Three: Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier() #create object
forest.fit(x_train, y_train) #train the object

forest_pred = forest.predict(x_test)
forest_accuracy = accuracy_score(y_test, forest_pred) #calculate the accuracy score

print("Accuracy Score of Random Forest Algorithm is: {:0.4f}%".format(forest_accuracy*100))
```

> **Feature Importance** refers to techniques that calculate a score for all the input features for a given model ‚Äî the scores simply represent the ‚Äúimportance‚Äù of each feature. A higher score means that the specific feature will have a larger effect on the model that is being used to predict a certain variable.



```python
forest = RandomForestClassifier() #create object
forest.fit(x_train, y_train) #train the object
feature_scores = pd.Series(forest.feature_importances_, index=x_train.columns).sort_values(ascending=False)
feature_scores.plot(kind='bar', figsize=(20,5), title='Feature Importance')
plt.show()
```

```python
x_train = x_train.loc[:, ['Customer_Age', 'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Amt_Chng_Q4_Q1', 'Total_Ct_Chng_Q4_Q1', 'Credit_Limit',
                          'Months_Inactive_12_mon', 'Months_on_book', 'Total_Relationship_Count', 'Total_Revolving_Bal']]
x_test = x_test.loc[:, ['Customer_Age', 'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Amt_Chng_Q4_Q1', 'Total_Ct_Chng_Q4_Q1', 'Credit_Limit',
                          'Months_Inactive_12_mon', 'Months_on_book', 'Total_Relationship_Count', 'Total_Revolving_Bal']]

forest = RandomForestClassifier() #create object
forest.fit(x_train, y_train) #train the object

forest_pred = forest.predict(x_test)
forest_accuracy = accuracy_score(y_test, forest_pred) #calculate the accuracy score

print("Accuracy Score of Random Forest Algorithm After feature_importances is: {:0.4f}%".format(forest_accuracy*100))
```
