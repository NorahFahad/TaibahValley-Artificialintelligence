# IMDB Movies Dataset 
This is data cleanning practice to the dataset [IMDB Movies Dataset](https://www.kaggle.com/datasets/harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows)
Top 1000 Movies by IMDB Rating


## Dataset columns description

| Column Name       | Description                                   |
| -----------------------------------------------------------|:-----------------------------------------------------------:| 
| Poster_Link       | Link of the poster that imdb using                | 
| Series_Title      | Name of the movie                                 |  
| Released_Year     | Year at which that movie released                 | 
| Certificate       | Certificate earned by that movie                  |   
| Runtime           | Total runtime of the movie                        |
| Genre             | Genre of the movie                                |
| IMDB_Rating       | Rating of the movie at IMDB site                  |
| Overview          | mini story/ summary                               |
| Meta_score        | Score earned by the movie                         |
| Director          | Name of the Director                              |
| Star1             | Name of the First actor star of the movie         |
| Star2             | Name of the Second actor star of the movie        |
| Star3             | Name of the Third actor star of the movie         |
| Star4             | Name of the Forth actor star of the movie         |
| No_of_Votes       | Total number of votes                             |
| Gross             | Money earned by that movie                        | 

information about some columns:

[Meta_score](https://www.imdb.com/list/ls051211184/): Metascore is considered the rating of a film.
Scores are assigned to movie's reviews of large group of the world's most respected critics, and weighted average are applied to summarize their opinions range.


[Genre](https://help.imdb.com/article/contribution/titles/genres/GZDRMS6R742JRGAG?ref_=helpms_helpart_inline#): A Genre is a category of artistic composition, characterized by similarities in form, style, or subject matter for a piece of content.

[Certificates](https://help.imdb.com/article/contribution/titles/certificates/GU757M8ZJ9ZPXB39?ref_=helpart_nav_27#): Certificate earned by that movie

## STEP 3: Clean Data
----------------

### **3.1: Filter unwanted column:-**
1. Poster_Link	
2. Overview
3. Runtime	

**will delete these columns**


```python
data.drop(columns=['Poster_Link', 'Overview', 'Runtime'], inplace=True) #drop columns
data.head() #dataset after delete 3 columns
```

```python
#dataset shape after delete 3 columns
data.shape
```

### **3.2: change column name:-**
*****
1.   Series_Title: better to change it to movie_name

```python
data.rename(columns={'Series_Title':'movie_name'}, inplace=True) #rename the column 
data.head() #after rename Series_Title
```

### **3.3: Split a Genre column:-**
*****
Splitting a Genre column into 3 column, main_genre which is the first category and genre2 and genre3

**3.3.1 split the Genre column**

```python
#split genre column into 3 column 
data[['main_genre', 'genre2', 'genre3']] = data['Genre'].str.split(', ', expand=True)
data.head() #after splitting
```

**3.3.2 delete the Genre after taking the values**

```python
data.drop("Genre", axis=1, inplace=True) #delete Genre column
data.head() #after drop Genre column
```

### **3.4: Handle missing data:-**
*****
1.   Certificate
2.   Meta_score
1.   Gross
4.   genre2
5.   genre3

```python
#sum of null values before
data.isnull().sum().to_frame()
```

**3.4.1 fill null values in Certificate column with 'Unknown'**

```python
#fill null values 
data['Certificate'].fillna('Unknown', inplace=True)
```

**3.4.2 fill null values in Meta_score column with the mean based on main_genre**

```python
#the mean of each category
data.groupby(['main_genre'])['Meta_score'].mean()
```

```python
print(data.loc[301,:]) #example for NaN in Meta_score column and the main_genre is Adventure 
```

```python
#fill null value in Meta_score column with the mean of each category
data['Meta_score'] = data.groupby(['main_genre'])['Meta_score'].transform(lambda x: x.fillna(x.mean()))
```

```python
print(data.loc[301,:]) #check if it fill with correct value (Adventure 78.437500)
```

```python
#check if there nulls
data[data['Meta_score'].isna()]
```

```python
#Fantasy movies does not have values in Meta_score column so it does not have mean value, will fill it with the mean of the second genre Horror 80.000000
data['Meta_score'].fillna(80.000, inplace=True)
```

```python
print(data.loc[321,:]) #check
```

**3.4.3 To fill in the null values in the Gross column with mean, must change the data type from Object to float to get the mean, in order to change from object to float must fix the syntax of the values by replace the ',' with ''**

```python
data['Gross'] = data['Gross'].str.replace(',', '') #fix the syntax
data = data.astype({"Gross":'float'}) #change gross data type
```

```python
pd.set_option('float_format', '{:f}'.format) #float format to print the full number
data.groupby(['main_genre'])['Gross'].mean() #get the mean of Gross column of each category
```

```python
print(data.loc[301,:]) #example for NaN in Gross column and the main_genre is Adventure 
```

```python
#fill null values in Gross column with the mean of each category
data['Gross'] = data.groupby(['main_genre'])['Gross'].transform(lambda x: x.fillna(x.mean()))
```

```python
print(data.loc[301,:]) #check if it fill with correct value (Adventure 86454989.901639)
```

```python
#check if there nulls
data[data['Gross'].isna()]
```

```python
#Fantasy movies does not have values in Gross column so it does not have mean value, will fill it with the mean of the second genre Horror 73585773.300000
data['Gross'].fillna(73585773.300000, inplace=True)
```

```python
print(data.loc[321,:]) #check
```

**3.4.4 will fill nulls  in this 2 column with 'None' means there is no second or third genre**

```python
#fill nulls with 'None' 
data.fillna({'genre2':'None','genre3':'None'}, inplace=True)
```

```python
#sum of null values after
data.isnull().sum().to_frame()
```

```python
#numric data description
data.describe()
```
